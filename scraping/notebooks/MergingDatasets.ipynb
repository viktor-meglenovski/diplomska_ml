```python
import pandas as pd
import numpy as np
```

```python
dfs=list()
dates=('2023-07-30','2023-08-15','2023-08-24')
merged_df=pd.read_excel(f'..\\data\\all_data_{dates[0]}.xlsx')
merged_df.rename(columns={"regular_price":dates[0]},inplace=True)
for date in dates[1:]:
    df=pd.read_excel(f'..\\data\\all_data_{date}.xlsx')
    df.rename(columns={"regular_price":date},inplace=True)
    df=df[['link',date]]
    merged_df = merged_df.merge(df, on="link", how="inner")
```

```python
columns=['name','link','image','category','store']
columns+=dates
merged_df=merged_df[columns]
```
